# ML TIME!

```python
core_weather["target"] = core_weather.shift(-1)["temp_max"] # pulls temp max down 1 row - "tomorrows weather"
```

```python
core_weather
```

```python
core_weather = core_weather.iloc[:-1,:].copy() # deletes last row with missing target value
```

```python
core_weather
```

```python
from sklearn.linear_model import Ridge # imports scikit linear module and puts into reg

reg = Ridge(alpha=.1) 
```

```python
predictors = ["precip", "snow", "snow_depth", "temp_max", "temp_min"]
```

```python
train = core_weather.loc[:"2020-12-31"] # training dates are 2020-12-31 and past
```

```python
test = core_weather.loc["2021-01-01":] # testing dates are 2021-01-01 and future
```

```python
reg.fit(train[predictors], train["target"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(test["target"], predictions)
```

```python
combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1) # puts predictions side by side actual data
combined.columns = ["actual", "predictions"]
```

```python
combined
```

```python
combined.plot()
```

```python
reg.coef_ # precip, snow, snow_depth have a much heavier influence compared to temp_max, temp_min
```

```python
def create_predictions(predictors, core_weather, reg): # puts all previously done methods into a function that will create predictions
    train = core_weather.loc[:"2020-12-31"] 
    test = core_weather.loc["2021-01-01":] 
    reg.fit(train[predictors], train["target"])
    predictions = reg.predict(test[predictors])
    error = mean_absolute_error(test["target"], predictions)
    combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1) 
    combined.columns = ["actual", "predictions"]
    return error, combined
```

# Add more predictors

```python
core_weather["month_max"] = core_weather["temp_max"].rolling(30).mean() # gets rolling mean and assigns to month_max
```

```python
core_weather
```

```python
core_weather = core_weather.iloc[30:,:].copy() # removes first 30 cols
```

```python
core_weather
```

```python
core_weather["month_day_max"] = core_weather["month_max"] / core_weather["temp_max"]
```

```python
core_weather["max_min"] = core_weather["temp_max"] / core_weather["temp_min"]
```

```python
predictors = ["precip", "snow", "snow_depth", "temp_max", "temp_min", "month_max", "month_day_max", "max_min"]
```

```python
error, combined = create_predictions(predictors, core_weather, reg)
```

```python
error # ~ avg 4.959 degrees F off
```

```python
combined.plot()
```

# Adding more predictors

```python
# making a monthly avg & day of year avg that will add and average as days pass
core_weather['monthly_avg'] = core_weather.temp_max.groupby(core_weather.index.month).transform('mean')
core_weather["day_of_year_avg"] = core_weather.temp_max.groupby(core_weather.index.day_of_year).transform('mean')

```

```python
predictors = predictors + ["monthly_avg", "day_of_year_avg"]
```

```python
error, combined = create_predictions(predictors, core_weather, reg)
```
